```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("pyspark_test") \
    .getOrCreate()
```

```python
spark
```

```python
data = [
    ("Alice", 100),
    ("Bob", 200),
    ("Charlie", 300)
]

df = spark.createDataFrame(data, ["name", "amount"])
df.show()
```
